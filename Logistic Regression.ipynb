```python
x= pd.DataFrame([[1,2],[3,4],[5,6]],columns=['x1','x2'])
x=np.array(x)
y=np.array([0,0,1]).reshape(3,1)
w=np.random.uniform(0,1,(2,1))

def sigmoid (x):
    return np.exp(x)/(1+np.exp(x))

def compute_cost(x,y,w):
    p=sigmoid(x@w)
    epsilon=1e-5
    
    return -(y.T@np.log(p+epsilon)+(1-y.T)@np.log(1-p+epsilon))

def gradient_decent(steps,x,y,w,learning_rate):
    cost=[]
    for i in range(steps):
        predict=sigmoid(x@w)
        w = w + learning_rate*(x.T@(y-predict))
        cost.append(compute_cost(x,y,w))
#         display(w,learning_rate*(x.T@(y-predict)))
    return (cost,w,predict)
        
        
```
